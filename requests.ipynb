```python
import requests

url = "https://restcountries.com/v3.1/all"

try:
    response = requests.get(url)

    # Print status code
    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries:", len(data))
    else:
        print("Error: Failed to fetch data from API")

except requests.exceptions.RequestException as e:
    print("Request Error:", e)
```

```python
import pandas as pd

# Sample data for countries
sample_data = [
    {"common_name": "United States", "population": 331000000, "region": "Americas", "area": 9834000},
    {"common_name": "Canada", "population": 38000000, "region": "Americas", "area": 9985000},
    {"common_name": "Germany", "population": 83000000, "region": "Europe", "area": 357000},
    {"common_name": "Japan", "population": 126000000, "region": "Asia", "area": 378000},
    {"common_name": "Australia", "population": 25000000, "region": "Oceania", "area": 7692000}
]

# Create DataFrame from sample data
df = pd.DataFrame(sample_data)

print("Sample DataFrame 'df' created successfully.")
display(df.head())
print("DataFrame '10' contains the following columns:")

display(df.head(10))
```

```python

region_population = df.groupby("region")["population"].sum()
max_region = region_population.idxmax()

# largest area
largest_country = df.loc[df["area"].idxmax(), "common_name"]

#
print("\nSummary Results:")
print("Region with highest total population:", max_region)
print("Country with the largest area:", largest_country)
```
